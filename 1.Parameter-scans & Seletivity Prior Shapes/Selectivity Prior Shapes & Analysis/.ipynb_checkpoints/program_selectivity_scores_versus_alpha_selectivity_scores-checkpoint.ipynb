```python
import pandas as pd
PKIS2=pd.read_excel("PKIS2_dataset.xlsx")
isel_scores_700 = pd.read_csv("inhibitor_selectivity_scores_mu700_030422.csv")
isel_scores_200 = pd.read_csv("inhibitor_selectivity_scores_mu200_030422.csv")

print(PKIS2)
print(isel_scores_700)
print(isel_scores_200)
```

```python
import numpy as np
import statistics

#specificity factor: a = Ro*L / ((Ri*L) for all other R) 
R_dict = {}
for i in range(0, PKIS2.shape[0]-1):
    R_scores = []
    inhibitor_name = PKIS2.iloc[i,1]
    for j in range(7,PKIS2.shape[1]):
        total_affinity = (PKIS2.iloc[i,7:]).sum(axis=0)
        local_affinity = PKIS2.iloc[i,j]
        R = local_affinity/total_affinity
        R_scores.append(R)
    R_dict[inhibitor_name] = R_scores
#print(R_dict)
```

```python
#Compute the correlations for mu=700 and mu=200 against specificity factors for every inhibitor

cor_v_200 = []
cor_v_700 = []

for inhibitor in R_dict:
    
    cor_200 = None
    cor_700 = None
    
    R_scores = R_dict[inhibitor]
    
    for i in range(0,isel_scores_200.shape[0]):
        nan_indexes = []
        if isel_scores_200.iloc[i,0] == inhibitor:
            scores_200 = isel_scores_200.iloc[i,1:]
            scores_200.astype('float32')
            scores_200 = scores_200.to_numpy()
            for j in range(0,len(scores_200)):
                if np.isnan(scores_200[j]):
                    nan_indexes.append(j)
            filtered_scores = [scores_200[i] for i in range(0,len(scores_200)) if i not in nan_indexes]
            filtered_Rscores = [R_scores[i] for i in range(0,len(R_scores)) if i not in nan_indexes]
            cor = np.corrcoef(filtered_Rscores, filtered_scores)
            cor_200 = cor[0][1]
            
    for i in range(0,isel_scores_700.shape[0]):
        nan_indexes = []
        if isel_scores_700.iloc[i,0] == inhibitor:
            scores_700 = isel_scores_700.iloc[i,1:]
            scores_700.astype('float32')
            scores_700 = scores_700.to_numpy()
            for j in range(0,len(scores_700)):
                if np.isnan(scores_700[j]):
                    nan_indexes.append(j)
            filtered_scores = [scores_700[i] for i in range(0,len(scores_700)) if i not in nan_indexes]
            filtered_Rscores = [R_scores[i] for i in range(0,len(R_scores)) if i not in nan_indexes]
            cor = np.corrcoef(filtered_Rscores, filtered_scores)
            cor_700 = cor[0][1]
            
    if cor_200 != None and cor_700 != None:
        cor_v_200.append(cor_200)
        cor_v_700.append(cor_700)
```

```python
import matplotlib.pyplot as plt

#from https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_hist.html#sphx-glr-gallery-lines-bars-and-markers-scatter-hist-py
def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    ax.scatter(x, y)
    ax.set_xlabel('Corrrelation coefficient v. mu=200 selectivity scores')
    ax.set_ylabel('Correlation coefficient v. mu=700 selectivity scores')

    # now determine nice limits by hand:
    binwidth = 0.0005
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(0.965, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
    
left, width = 1, 3
bottom, height = 1, 3
spacing = 0

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 1]
rect_histy = [left + width + spacing, bottom, 1, height]

# start with a square Figure
fig = plt.figure(figsize=(1.5, 1.5))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(cor_v_200, cor_v_700, ax, ax_histx, ax_histy)

ax.set_title('Correlation between Inhibitor Specificity Factors and Selectivity Scores',x=0.65, y=1.35, pad=0)

plt.show()
```
