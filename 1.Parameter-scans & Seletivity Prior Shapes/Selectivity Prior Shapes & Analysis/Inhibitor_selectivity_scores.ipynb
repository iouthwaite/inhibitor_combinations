```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import os

def import_from_excel(e_file, filename):
    kinase_targets = e_file['kinase target'].tolist()
    JS_scores = e_file['JS Distance Score'].tolist()
    inhibitors = e_file['inhibitors'].tolist()
    mu = int(filename.split('_')[-1].split('.')[0])
    return [mu, kinase_targets, JS_scores, inhibitors]

test_results_200 = []
test_results_700 = []

for filename in os.listdir(os.getcwd()+'/representative_results/'):
    if filename[0] != '.' and filename[0] != '~':
        f = './representative_results/'+str(filename)
        open_file = pd.read_excel(f, '1')
        res = import_from_excel(open_file, filename)
        if res[0] == 200:
            test_results_200 = res
        if res[0] == 700:
            test_results_700 = res
        
#set up the list of all kinase names for our result matrix
PKIS2_data=pd.read_excel("PKIS2_dataset.xlsx")
kinase_names = list(PKIS2_data.columns.values)[7:]
num_kinases = len(kinase_names)

print(test_results_200)
```

```python
#get the scores of every inhibitor against every kinase 
inhibitor_names = []

result_array = []

KT_200 = test_results_200[1]
I_200 = test_results_200[3]
JSD_200 = test_results_200[2]

#set up the array
unique_inhibitor_names = list(set(I_200))
#print(unique_inhibitor_names)
for h in range(0, len(unique_inhibitor_names)):
    result_array.append([])
    i_name = unique_inhibitor_names[h][2:-2]
    inhibitor_names.append(i_name)
    for j in range(num_kinases):
        result_array[h].append(None)

#fill the array - score is the JSD
for x in range(0, len(KT_200)):
    I = I_200[x][2:-2]
    KT = KT_200[x]
    JSD = JSD_200[x]
    score = JSD
    I_index = inhibitor_names.index(I)
    KT_index = kinase_names.index(KT)
    print(I_index, KT_index)
    result_array[I_index][KT_index] = score
    
print(result_array)
    
res_arr = np.array(result_array, dtype=object)

#same thing, for mu=700

KT_700 = test_results_700[1]
I_700 = test_results_700[3]
JSD_700 = test_results_700[2]

inhibitor_names2 = []
result_array2 = []

#set up the array
unique_inhibitor_names2 = list(set(I_700))
for h in range(0, len(unique_inhibitor_names2)):
    result_array2.append([])
    i_name = unique_inhibitor_names2[h][2:-2]
    inhibitor_names2.append(i_name)
    for j in range(num_kinases):
        result_array2[h].append(None)

#fill the array - score is the JSD (a measure of the off-target selectivity) * the on-target % inhibition
for x in range(0, len(KT_700)):
    I = I_700[x][2:-2]
    KT = KT_700[x]
    JSD = JSD_700[x]
    score = JSD
    I_index = inhibitor_names2.index(I)
    KT_index = kinase_names.index(KT)
    result_array2[I_index][KT_index] = score
    
res_arr2 = np.array(result_array2, dtype=object)

#print(res_arr)
```

```python
#  now that we have the JSD score array, normalize the score for each inhibitor against 
#  a) each inhibitor's maximum or 
#  b) each kinase's maximum score

'''
#A - how selective is an inhibitor for a kinase relative to its activity against other kinases?
scores_norm_imax = np.copy(res_arr)
for i in range(0,len(scores_norm_imax)):
    scores = [x for x in scores_norm_imax[i] if x != None]
    max_score = np.max(scores)
    min_score = np.min(scores)
    for j in range(0, len(scores_norm_imax[i])):
        if scores_norm_imax[i][j] != None:
            scores_norm_imax[i][j] = (scores_norm_imax[i][j] - min_score) / (max_score - min_score)
'''

#B - how selective is an inhibitor to a kinase relative to other inhibitors that also target that kinase? At mu=200
scores_norm_kmax = np.copy(res_arr)
for j in range(0,len(scores_norm_kmax[0])):
    col = scores_norm_kmax[:, j]
    scores = [x for x in col if x != None]
    if len(scores) > 0:
        max_score = np.max(scores)
        min_score = np.min(scores)
        for i in range(0, len(scores_norm_kmax)):
            if scores_norm_kmax[i][j] != None:
                scores_norm_kmax[i][j] = (scores_norm_kmax[i][j] - min_score) / (max_score - min_score)

#B - how selective is an inhibitor to a kinase relative to other inhibitors that also target that kinase? At mu=700
scores_norm_kmax2 = np.copy(res_arr2)
for j in range(0,len(scores_norm_kmax2[0])):
    col = scores_norm_kmax2[:, j]
    scores = [x for x in col if x != None]
    if len(scores) > 0:
        max_score = np.max(scores)
        min_score = np.min(scores)
        for i in range(0, len(scores_norm_kmax2)):
            if scores_norm_kmax2[i][j] != None:
                scores_norm_kmax2[i][j] = (scores_norm_kmax2[i][j] - min_score) / (max_score - min_score)
```

```python
#output the results to a table
import pandas as pd

result_list = []
header = ['Inhibitor'] + kinase_names
#result_list.append(header)
for i in range(0, len(scores_norm_kmax)):
    inhibitor_name = inhibitor_names[i]
    data = [inhibitor_name] + list(scores_norm_kmax[i])
    result_list.append(data)
df = pd.DataFrame(result_list, columns=header)
print(df)
df.to_csv("inhibitor_selectivity_scores_mu200.csv",index=False)

result_list2 = []
header = ['Inhibitor'] + kinase_names
#result_list.append(header)
for i in range(0, len(scores_norm_kmax2)):
    inhibitor_name = inhibitor_names2[i]
    data = [inhibitor_name] + list(scores_norm_kmax2[i])
    result_list2.append(data)
df2 = pd.DataFrame(result_list2, columns=header)
print(df2)
df2.to_csv("inhibitor_selectivity_scores_mu700.csv",index=False)
```
